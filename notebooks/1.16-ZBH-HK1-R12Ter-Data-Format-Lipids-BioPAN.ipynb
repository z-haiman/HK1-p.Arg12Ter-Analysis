Currently, BioPAN:

* Does not analyse oxidised lipids.
* Consider by default the sphingoid bases with 18 carbons and 0 (dhCer/dhSM) or 1 double bond (Cer/SM) for the Sphingolipids. For example: it means that the lipids molecular species ‘Cer(24:0)’ corresponds to ‘Cer(d18:1/24:0)’ and ‘dhCer(20:4)’ corresponds to ‘Cer(d18:0/20:4)’ (same for SM/dhSM).


```python
df_biopan = df.copy()
df_classes = df_biopan["Omics Name"].str.extract(r"(\S+)\((\S+)\)")
# Identify if lipid is an ether lipid
df_classes[1] = df_classes.iloc[:, 1].str.extract(r"(O|P)-")
df_classes = df_classes.apply(
    lambda x: x[0] if len(x.dropna().values) <= 1 else f"{x[1]}-{x[0]}", axis=1
)
df_biopan["BioPAN Lipid Subclass"] = df_classes.fillna(df_biopan["Omics Lipid Class"])
biopan_class_map = {
    "AcCa": "",
    # 'Cer': 'Cer', # Needs extra processing
    "Ch": "ST",
    "ChE": "ST",
    # 'CL': 'CL',
    "O-CL": "",
    # 'DG': 'DG',
    # 'O-DG': 'O-DG',
    "P-DG": "",
    # 'FA': 'FA',
    "GD3": "",
    "GM3": "",
    "Hex1Cer": "Hex-Cer",  # Needs extra processing
    "Hex2Cer": "Hex-Cer",  # Needs extra processing
    "Hex3Cer": "Hex-Cer",  # Needs extra processing
    # 'LPC': 'LPC',
    # 'O-LPC': 'O-LPC',
    # 'P-LPC': 'P-LPC',
    # 'LPE': 'LPE',
    # 'O-LPE': 'O-LPE',
    # 'LPI': 'LPI',
    # 'LPS': 'LPS',
    # 'PA': 'PA',
    # 'PC': 'PC',
    # 'O-PC': 'O-PC',
    # 'P-PC': 'P-PC',
    # 'PE': 'PE',
    # 'O-PE': 'O-PE',
    # 'PG': 'PG',
    "O-PG": "",
    "PI": "PI",
    "PS": "PS",
    "O-PS": "",
    "P-PS": "",
    "SM": "SM",
    "SPH": "SPB",
    "SPHP": "SPBP",
    "ST": "ST",
    "TG": "TG",
    "O-TG": "",
    "P-TG": "",
}
df_biopan["BioPAN Lipid Subclass"] = df_biopan["BioPAN Lipid Subclass"].replace(biopan_class_map)

# Process ceramides
df_further_process = df_biopan[df_biopan["BioPAN Lipid Subclass"] == "Cer"]
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"Cer\(d18:0|Cer\(d18:1"),
    "BioPAN Lipid Subclass",
] = ""
df_further_process.loc[
    df_further_process["Omics Name"].str.startswith("Cer(d18:0"), "BioPAN Lipid Subclass"
] = "dhCer"
df_biopan.loc[df_further_process.index, "BioPAN Lipid Subclass"] = df_further_process

df_further_process = df_biopan[df_biopan["BioPAN Lipid Subclass"] == "Hex-Cer"]
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"Hex1Cer\(d18:1"), "BioPAN Lipid Subclass"
] = ""
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"Hex2Cer\(d18:1"), "BioPAN Lipid Subclass"
] = ""
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"Hex3Cer\(d18:1"), "BioPAN Lipid Subclass"
] = ""
df_further_process.loc[
    df_further_process["Omics Name"].str.contains(r"Hex1Cer\(d18:1"), "BioPAN Lipid Subclass"
] = "Hex-Cer"
df_further_process.loc[
    df_further_process["Omics Name"].str.contains(r"Hex2Cer\(d18:1"), "BioPAN Lipid Subclass"
] = "Hex-Cer"
df_further_process.loc[
    df_further_process["Omics Name"].str.contains(r"Hex3Cer\(d18:1"), "BioPAN Lipid Subclass"
] = "Hex-Cer"
df_biopan.loc[df_further_process.index, "BioPAN Lipid Subclass"] = df_further_process

# Process sphingolipids
df_further_process = df_biopan[df_biopan["BioPAN Lipid Subclass"] == "SM"]
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"SM\(d18:0|SM\(d18:1"), "BioPAN Lipid Subclass"
] = ""
df_further_process.loc[
    df_further_process["Omics Name"].str.startswith("SM(d18:0"), "BioPAN Lipid Subclass"
] = "dhSM"
df_biopan.loc[df_further_process.index, "BioPAN Lipid Subclass"] = df_further_process

df_further_process = df_biopan[df_biopan["BioPAN Lipid Subclass"] == "SPB"]
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"SPH\(d18:0|SPH\(d18:1"),
    "BioPAN Lipid Subclass",
] = ""
df_further_process.loc[
    df_further_process["Omics Name"].str.startswith("SPH(d18:0"), "BioPAN Lipid Subclass"
] = "dhSPB"
df_further_process.loc[
    df_further_process["Omics Name"].str.startswith("SPH(d18:1"), "BioPAN Lipid Subclass"
] = "SPB"
df_biopan.loc[df_further_process.index, "BioPAN Lipid Subclass"] = df_further_process

# Process sphingolipids
df_further_process = df_biopan[df_biopan["BioPAN Lipid Subclass"] == "SPBP"]
df_further_process.loc[
    ~df_further_process["Omics Name"].str.contains(r"SPHP\(d18:0|SPHP\(d18:1"),
    "BioPAN Lipid Subclass",
] = ""
df_further_process.loc[
    df_further_process["Omics Name"].str.startswith("SPHP(d18:0"), "BioPAN Lipid Subclass"
] = "dhSPBP"
df_further_process.loc[
    df_further_process["Omics Name"].str.startswith("SPHP(d18:1"), "BioPAN Lipid Subclass"
] = "SPBP"
df_biopan.loc[df_further_process.index, "BioPAN Lipid Subclass"] = df_further_process


df_biopan["BioPAN Lipid Subclass"] = df_biopan["BioPAN Lipid Subclass"].replace("", pd.NA)
df_biopan = df_biopan.dropna(subset="BioPAN Lipid Subclass", axis=0)
df_biopan
```

### Determine sum composition of lipid

```python
df_biopan.loc[:, "BioPAN Lipid Sum Composition"] = pd.NA
# Process ceramides
df_further_process = df_biopan[df_biopan["Omics Lipid Class"].str.endswith("Cer")]["Omics Name"]
df_further_process = df_further_process.str.replace("Cer(d18:0_", "dhCer(")
df_further_process = df_further_process.str.replace("Cer(d18:1_", "Cer(")
df_further_process = df_further_process.str.replace("Hex1", "Hex-")
df_further_process = df_further_process.str.replace("Hex2", "Hex-")
df_further_process = df_further_process.str.replace("Hex3", "Hex-")
df_biopan.loc[
    df_biopan[df_biopan["Omics Lipid Class"].str.endswith("Cer")].index,
    "BioPAN Lipid Sum Composition",
] = df_further_process

# Process sphingolipids
df_further_process = df_biopan[df_biopan["Omics Lipid Class"].str.endswith("SM")]["Omics Name"]
df_further_process = df_further_process.str.replace("SM(d18:0_", "dhSM(")
df_further_process = df_further_process.str.replace("SM(d18:1_", "SM(")
df_biopan.loc[
    df_biopan[df_biopan["Omics Lipid Class"].str.endswith("SM")].index,
    "BioPAN Lipid Sum Composition",
] = df_further_process

df_further_process = df_biopan[df_biopan["Omics Lipid Class"] == "SPH"]["Omics Name"]
df_further_process = df_further_process.str.replace("SPH(d18:0)", "dhSPB")
df_further_process = df_further_process.str.replace("SPH(d18:1)", "SPB")
df_biopan.loc[
    df_biopan[df_biopan["Omics Lipid Class"] == "SPH"].index, "BioPAN Lipid Sum Composition"
] = df_further_process

df_further_process = df_biopan[df_biopan["Omics Lipid Class"] == "SPHP"]["Omics Name"]
df_further_process = df_further_process.str.replace("SPHP(d18:0)", "dhSPBP")
df_further_process = df_further_process.str.replace("SPHP(d18:1)", "SPBP")
df_biopan.loc[
    df_biopan[df_biopan["Omics Lipid Class"] == "SPHP"].index, "BioPAN Lipid Sum Composition"
] = df_further_process

df_further_process = df_biopan[df_biopan["Omics Lipid Class"].str.startswith("Ch")]["Omics Name"]
df_further_process = df_further_process.str.replace("ChE", "ST")
df_further_process = df_further_process.str.replace("Ch", "ST")
df_biopan.loc[
    df_biopan[df_biopan["Omics Lipid Class"].str.startswith("Ch")].index,
    "BioPAN Lipid Sum Composition",
] = df_further_process

for lc in [
    "FA",
    "LPC",
    "LPE",
    "LPI",
    "LPS",
    "PA",
    "DG",
    "TG",
    "PC",
    "PE",
    "PS",
    "PI",
    "CL",
    "PG",
    "ST",
]:
    df_biopan.loc[
        df_biopan.loc[df_biopan["Omics Lipid Class"] == lc].index, "BioPAN Lipid Sum Composition"
    ] = df_biopan.loc[df_biopan["Omics Lipid Class"] == lc, "Omics Name"]
    if lc in {"DG", "LPC", "PC", "LPE", "PE"}:
        df_biopan.loc[
            df_biopan.loc[df_biopan["Omics Lipid Class"] == lc].index,
            "BioPAN Lipid Sum Composition",
        ] = df_biopan.loc[
            df_biopan.loc[df_biopan["Omics Lipid Class"] == lc].index,
            "BioPAN Lipid Sum Composition",
        ].str.replace(
            f"{lc}(O-", f"O-{lc}("
        )
    if lc in {"LPC", "PC"}:
        df_biopan.loc[
            df_biopan.loc[df_biopan["Omics Lipid Class"] == lc].index,
            "BioPAN Lipid Sum Composition",
        ] = df_biopan.loc[
            df_biopan.loc[df_biopan["Omics Lipid Class"] == lc].index,
            "BioPAN Lipid Sum Composition",
        ].str.replace(
            f"{lc}(P-", f"P-{lc}("
        )

df_sum_comp = (
    df_biopan["BioPAN Lipid Sum Composition"].str.extractall(r"\((\S+)\)").droplevel(level="match")
)
df_sum_comp["components"] = df_sum_comp[0].str.split("_")
df_sum_comp["C"] = df_sum_comp["components"].apply(
    lambda values: str(sum([int(x.strip("d").split(":")[0]) for x in values]))
)
df_sum_comp["DB"] = df_sum_comp["components"].apply(
    lambda values: str(sum([int(x.strip("d").split(":")[1]) for x in values]))
)
df_sum_comp["O"] = df_sum_comp["components"].apply(
    lambda values: list(set(["O2" if x.startswith("d") else "" for x in values]))[0]
)
df_sum_comp["C:DB"] = df_sum_comp[["C", "DB"]].apply(lambda values: ":".join(values), axis=1)
df_sum_comp["Composition"] = df_sum_comp[["C:DB", "O"]].apply(
    lambda values: ";".join([x for x in values if x]), axis=1
)
df_sum_comp["Class"] = df_biopan.loc[
    df_biopan.index.isin(df_sum_comp.index), "BioPAN Lipid Subclass"
]
df_biopan.loc[df_biopan.index.isin(df_sum_comp.index), "BioPAN Lipid Sum Composition"] = (
    df_sum_comp[["Class", "Composition"]].apply(
        lambda x: f"{x['Class']}({x['Composition']})", axis=1
    )
)
df_biopan["Recognized by BioPAN"] = "Yes"
df_biopan = df_biopan.drop("Omics Lipid Class", axis=1)

df_biopan = pd.concat(
    (df_biopan, df[~df["Omics Name"].isin(df_biopan["Omics Name"].values)]["Omics Name"]), axis=0
)
df_biopan["Recognized by BioPAN"] = df_biopan["Recognized by BioPAN"].fillna("No")
df_biopan = df_biopan.fillna("").sort_values(
    ["Recognized by BioPAN", "BioPAN Lipid Sum Composition"], ascending=[False, True]
)
```

## Apply to lipid intensity data
### Filter for lipids

```python
source_types = ["RBC", "Plasma"]

# Biopan performs sum normalization on samples before scaling, use raw intensity data
norm_str = "none-none-none"

# Input data args
input_value_type = "Intensities"
input_dirpath = PROCESSED_DATA_DIR / "HK1-R12Ter" / "Normalized" / norm_str

replacement_map = df_biopan.set_index("Omics Name")["BioPAN Lipid Sum Composition"].to_dict()
index_cols = [key for key in [sample_key, group_key] if key]
for source_type in source_types:
    data_type = "CombinedLipids"
    logger.debug(f"Processing data for '{source_type}-{data_type}'...")
    filename = make_filename(source_type, data_type, input_value_type)
    df = load_dataframe_from_file(input_dirpath / filename, index_col=index_cols)

    # Remove lipids that cannot map to BioPan
    df = df.loc[:, df.columns.isin(list(df_biopan["Omics Name"]))].T
    # Remove phenotype groups from index
    df = df.droplevel(1, axis=1)

    df = df.rename(replacement_map, axis=0)
    df = df.rename(
        {col_name: col_name.split("_", maxsplit=1)[-1] for col_name in df.columns}, axis=1
    )
    df = df.loc[[x for x in df.index if x]].groupby(level=0).sum()

    data_type = "Lipids"
    filename = make_filename(source_type, data_type, output_value_type)
    save_dataframe_to_file(df, output_dirpath / filename, index=True)
    logger.info(f"Processed data for '{source_type}-{data_type}'.")
```
